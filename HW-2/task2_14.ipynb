```python
def a(n)
    sum = 0.0            # O(1)
    for i in range(n+1): # O(n)  
        sum += i         # O(1)    
    return sum           # O(1)
    
# T(n) = O(n)
```

```python
def b(n):
    sum = 0.0            # O(1)
    for i in range(n+1): # O(n)
        sum += i * i     # O(1)
    return sum           # O(1)
    
# T(n) = O(n)
```

```python
def c(a, n):
    sum = 1.0            # O(1)
    curr_pow = 1.0       # O(1)
    for i in range(n+1): # O(n)
        curr_pow *= a    # O(1)
        sum += curr_pow  # O(1)
    return sum           # O(1)
    
# T(n) = O(n)
```

```python
def d(n):
    sum = 0.0              # O(1)
    for i in range(n+1):   # O(n)
        if i == 0:         # O(1)
            sum += 1.0     # O(1)
            continue       # O(1)
            
        term = 1.0         # O(1)
        for k in range(i): # O(i)
            term *= i      # O(1)
        sum += term        # O(1)
        
    return sum             # O(1)
    
# When i=n: O(i) = O(n), thus
# T(n) = O(n) * O(n) = O(n^2)
```

```python
def e(n):
    res = 1.0                # O(1)
    for i in range(1, n+1):     # O(n)
        res *= 1.0 / (i + 1) # O(1)
    return res               # O(1)

# T(n) = O(n)
```

```python
def f(n):
    res = 1.0                       # O(1)
    curr_fac = 1.0                  # O(1)
    for i in range(1, n+1):         # O(n)
        curr_fac *= i               # O(1)
        res *= 1.0 / (curr_fac + 1) # O(1)
    return res                      # O(1)

# T(n) = O(n)
```

```python
def g(a, n):
    res = 1.0                            # O(1)
    curr_fac = 1.0                       # O(1)
    curr_pow = 1.0                       # O(1)
    for i in range(1, n+1):              # O(n)
        curr_pow *= a                    # O(1)
        curr_fac *= i                    # O(1)
        term = curr_pow / (curr_fac + 1) # O(1)
        res *= term                      # O(1)
    return res                           # O(1)

# T(n) = O(n)
```

```python
def h(n, m):
    res = 1.0                      # O(1)
    for i in range(1, n + 1):      # O(n)
        val_pow = 1.0              # O(1)
        for k in range(m):         # O(m) - const
            val_pow *= i           # O(1)
        res *= 1.0 / (1 + val_pow) # O(1)
    return res                     # O(1)

# T(n) = O(n) * O(m) = O(nm)
```

```python
def i(n):
    res = 1.0                      # O(1)
    for i in range(1, n + 1):      # O(n)
        val_pow = 1.0              # O(1)
        for k in range(i):         # O(i) - not const
            val_pow *= i           # O(1)
        res *= 1.0 / (1 + val_pow) # O(1)
    return res                     # O(1)

# T(n) = O(n) * O(n) = O(n^2)
```
