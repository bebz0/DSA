### a)
$$\begin{aligned}
T(n) &= T(n-1) + C \\
&= (T(n-2) + C) + C = T(n-2) + 2C \\
&= (T(n-3) + C) + 2C = T(n-3) + 3C \\
&\vdots \\
&= T(n-k) + kC
\end{aligned}$$$$k = n \implies n - k = 0$$$$T(n) = T(0) + nC = C_0 + nC = O(n)$$

### d)
$$\begin{aligned}
T(n) &= aT(n-a) + C \\
&= a(aT(n-2a) + C) + C = a^2 T(n-2a) + aC + C \\
&= a^2(aT(n-3a) + C) + aC + C = a^3 T(n-3a) + a^2C + aC + C \\
&\vdots \\
&= a^k T(n-ka) + C \sum_{i=0}^{k-1} a^i
\end{aligned}$$$$k \approx \frac{n}{a} \implies n - ka \approx 0$$$$T(n) \approx a^{n/a} T(0) + C \frac{a^{n/a} - 1}{a - 1} = O(a^{n/a})$$

### g)
$$\begin{aligned}
T(n) &= aT\left(\frac{n}{a}\right) + C \\
&= a\left(aT\left(\frac{n}{a^2}\right) + C\right) + C = a^2 T\left(\frac{n}{a^2}\right) + aC + C \\
&= a^2\left(aT\left(\frac{n}{a^3}\right) + C\right) + aC + C = a^3 T\left(\frac{n}{a^3}\right) + a^2C + aC + C \\
&\vdots \\
&= a^k T\left(\frac{n}{a^k}\right) + C \sum_{i=0}^{k-1} a^i
\end{aligned}$$$$\frac{n}{a^k} = 1 \implies a^k = n \implies k = \log_a n$$$$\begin{aligned}
T(n) &= n T(1) + C \frac{a^k - 1}{a - 1} \\
&= n \cdot C + C \frac{n - 1}{a - 1} \\
&= O(n)
\end{aligned}$$

### h)
$$\begin{aligned}
T(n) &= aT\left(\frac{n}{a}\right) + Cn \\
&= a\left(aT\left(\frac{n}{a^2}\right) + C\frac{n}{a}\right) + Cn \\
&= a^2 T\left(\frac{n}{a^2}\right) + Cn + Cn = a^2 T\left(\frac{n}{a^2}\right) + 2Cn \\
&= a^2\left(aT\left(\frac{n}{a^3}\right) + C\frac{n}{a^2}\right) + 2Cn \\
&= a^3 T\left(\frac{n}{a^3}\right) + Cn + 2Cn = a^3 T\left(\frac{n}{a^3}\right) + 3Cn \\
&\vdots \\
&= a^k T\left(\frac{n}{a^k}\right) + kCn
\end{aligned}$$$$\frac{n}{a^k} = 1 \implies a^k = n \implies k = \log_a n$$$$\begin{aligned}
T(n) &= n T(1) + (\log_a n) Cn \\
&= O(n) + O(n \log n) \\
&= O(n \log n)
\end{aligned}$$
