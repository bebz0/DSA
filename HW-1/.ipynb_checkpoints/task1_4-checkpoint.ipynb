### a)

$$
T(n) = \begin{cases} 1, & n \le a; \\ T(n-a) + 1, & n > a. \end{cases}
$$



$$
T(n) = T(n-a) + 1 = T(n-2a) + 2 = \dots = T(n - k \cdot a) + k
$$

$$n - k \cdot a = 0 \implies k = \frac{n}{a}$$

$$
T(n) = T(00) + \frac{n}{a} = 1 + \frac{n}{a}
$$

$T(n) = \frac{n}{a} + 1$


### b)

$$
T(n) = \begin{cases} 1, & n = 0; \\ T(n-1) + 2^n, & n \ge 1. \end{cases}
$$

$$
T(n) = T(n-1) + 2^n = T(n-2) + 2^{n-1} + 2^n = \dots = T(0) + \sum_{k=1}^{n} 2^k
$$


$$
T(n) = \sum_{i=0}^{n} 2^i = \frac{2^{n+1} - 1}{2 - 1} = 2^{n+1} - 1
$$

$T(n) = 2^{n+1} - 1$

### c)

$$
T(n) = \begin{cases} 1, & n = 1; \\ 2T(\lfloor n/2 \rfloor) + 1, & n \ge 2. \end{cases}
$$



$$
T(n) = 2T(n/2) + 1 = 2(2T(n/4) + 1) + 1 = 4T(n/4) + 2 + 1 == \dots
$$

$$
T(n) = 2^k T(n/2^k) + \sum_{i=0}^{k-1} 2^i = 2^k T(n/2^k) + (2^k - 1)
$$


$$
T(n) = n \cdot T(1) + (n - 1) = n \cdot 1 + n - 1 = 2n - 1
$$

$T(n) = 2n - 1$

### d)

$$
T(n) = \begin{cases} 1, & n = 1; \\ aT(\lfloor n/a \rfloor) + n, & n \ge 2, a \ge 2. \end{cases}
$$

$$
T(n) = aT\left(\frac{n}{a}\right) + n = a\left[aT\left(\frac{n}{a^2}\right) + \frac{n}{a}\right] + n = a^2 T\left(\frac{n}{a^2}\right) + 2n = \dots
$$

$$
T(n) = a^k T\left(\frac{n}{a^k}\right) + k \cdot n
$$


$$
T(n) = n \cdot T(1) + n \cdot \log_a n = n \cdot 1 + n \log_a n
$$

$T(n) = n(1 + \log_a n)$
