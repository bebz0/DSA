```python
def a(n,k):
    k += 1       # 4
    i = n        # 2
    while i > 0: # 3 * (n + 1) 
        i -= 1   # 4 * n
    return None
'''
n=0: 1, 0 
n=1: 2, 1
n=2: 3, 2
n=3: 4, 3
...
'''
#T(n) = 4 + 2 + 3(n + 1) + 4n = 7n + 9 
```

### b)

```python
def b(n,k):
    i = n        # 2
    while i > 1: # 3 * (m + 1)
        k += 1   # 4 * m
        i //= 2  # 4 * m
    return None
'''
n=0: 1, 0
n=1: 1, 0
n=2: 2, 1
n=3: 2, 1
n=4: 3, 2
...
'''
# m = [log2(n)], n = 2 ** m
#T(n) = 2 + 3(m + 1) + 4m + 4m = 11m + 5 = 11[log2(n)] + 5
```

### c)

```python
def c(n,k):
    i = 0            # 2
    while i < n:     # 3 * (m + 1)
        j = 0        # 2 * m
        while j < n: # 3 * m * (m + 1)
            k += 1   # 4 * m * m
            j += 2   # 4 * m * m
        i += 2       # 4 * m
    return None
    
'''
n=0: 1, 0, 0, 0
n=1: 2, 1, 2, 1
n=2: 2, 1, 2, 1
n=3: 3, 2, 3, 2
n=4: 3, 2, 3, 2
n=5: 4, 3, 4, 3
...
'''

# m = [n / 2]
# T(n) = 2 + 3(m + 1) + 2m + 3m(m + 1) + 4m^2 + 4m^2 + 4m = 11m^2 + 12m + 5 = 11([n/2])^2 + 12[n/2] + 5
```

### d)

```python
def d(n,k):
    i = 0                # 2
    while i < n:         # 3 * (n + 1)
        j = 0            # 2 * n
        while j < i * i: # 5 * ( n + ((n-1)^2 + (n-2)^2 + ...  + 1^2 + 0^2)) = 5 * (n + S)
            k += 1       # 4 * S
            j += 1       # 4 * S
        i += 1           # 4 * n
    return None
    
'''
n=0: 1, 0, 0, 0
n=1: 2, 1, 1, 0
n=2: 3, 2, 2+1=3, 1
n=3: 4, 3, 3+5=8, 1+(5-1)=5
n=4: 5, 4, 8+10=18, 5+(10-1)=14
n=5: 6, 5, 18+17=35, 14+(17-1)=30
...
'''

# S = ((n-1)^2 + (n-2)^2 + ...  + 1^2 + 0^2) = (n * (n - 1) * (2 * n - 1)) / 6
# T(n) = 2 + 3 * (n + 1) + 2 * n + 5 * (n + S) + 4 * S + 4 * S + 4 * n = 13S + 14n + 5
```

### e)

```python
def e(n,k):
    i = 1            # 2
    while i < n:     # 3 * (m + 1)
        j = 1        # 2 * m
        while j < n: # 3 * m * (m + 1)
            k += 1   # 4 * m * m
            j *= 2   # 4 * m * m
        i *= 2       # 4 * m
    return None
    
'''
n=0: 1, 0, 0, 0
n=1: 1, 0, 0, 0
n=2: 2, 1, 2, 1
n=3: 3, 2, 6, 4
n=4: 3, 2, 6, 4
n=5: 4, 3, 12, 9
...
'''

# m = [log2(n)], n = 2 ** m
# T(n) = 2 + 3(m + 1) + 2m + 3m(m + 1) + 4m^2 + 4m^2 + 4m = 11m^2 + 8m + 5 = 11([log2(n)])^2 + 8[log2(n)] + 5 
```

### f)

```python
def f(n,k):
    i = 1            # 2
    while i < n:     # 3 * (m + 1)
        j = i        # 2 * m
        while j < n: # 3 * (S + m)
            k += 1   # 4 * S
            j *= 2   # 4 * S
        i *= 2       # 4 * m
    return None
    
'''
n=0: 1, 0, 0, 0
n=1: 1, 0, 0, 0
n=2: 2, 1, 2, 1
n=3: 3, 2, 5, 2+1=3
n=4: 3, 2, 5, 2+1=3
n=5: 4, 3, 9, 3+2+1=6
n=8: 4, 3, 9, 3+2+1=6
n=9: 5, 4, 14, 4+3+2+1=10
...
'''
# m = [log2(n)], n = 2 ** m
# S = m + (m-1) + (m-2) + ... + 1 = (m(m+1))/2
# T(n) = 2 + (3m + 3) + 2m + (3S + 3m) + 4S + 4S + 4m = 5.5([log2(n)])^2 + 17.5[log2(n)] + 5
```
